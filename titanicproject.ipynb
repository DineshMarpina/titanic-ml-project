```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("C:/Users/marpi/Downloads/ml/mldatasets/titanic/train.csv")
df=df.set_index("PassengerId")
df.head()
```

```python
df["Sex"].head()
df["Sex"]=df["Sex"].replace({"male":1,"female":0})
```

```python
df["Sex"].head()
```

```python
df["Embarked"].head()
```

```python
df["Embarked"]=df["Embarked"].replace({"S":0,"C":1,"Q":2})
df["Embarked"]=df["Embarked"].fillna(1)
df["Embarked"].head()
```

```python
df.info()
```

```python
df=df.drop(['Name',"Ticket","Cabin","SibSp","Parch"],axis=1)
```

```python
df["Age"]=df['Age'].fillna(df["Age"].mean())
df["Age"]=np.where((df["Age"]<=18),0,df["Age"])
df["Age"]=np.where((df["Age"]>18) & df["Age"]<=50,1,df["Age"])
df["Age"]=np.where((df["Age"]>50),2,df["Age"])
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
dft= pd.read_csv("C:/Users/marpi/Downloads/ml/mldatasets/titanic/test.csv")
```

```python
dft=dft.set_index("PassengerId")
dft.head()
```

```python
dft["Sex"].head()
dft["Sex"]=dft["Sex"].replace({"male":1,"female":0})

dft["Embarked"]=dft["Embarked"].replace({"S":0,"C":1,"Q":2})
dft["Embarked"]=dft["Embarked"].fillna(1)

dft=dft.drop(['Name',"Ticket","Cabin","SibSp","Parch"],axis=1)





dft["Age"]=dft['Age'].fillna(dft["Age"].mean())
dft["Age"]=np.where((dft["Age"]<=18),0,dft["Age"])
dft["Age"]=np.where((dft["Age"]>18) & dft["Age"]<=50,1,dft["Age"])
dft["Age"]=np.where((dft["Age"]>50),2,dft["Age"])

dft["Fare"]=dft["Fare"].fillna(dft["Fare"].mean())
```

```python
dft.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=["Survived"])
Y=df["Survived"]
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)

print("The accuracy of the model is :",model.score(x_test,y_test)*100)
pred=model.predict(dft)
```

```python
sub=pd.read_csv("C:/Users/marpi/Downloads/ml/mldatasets/titanic/gender_submission.csv")
sub["Survived"]=pred
sub.to_csv("sub lr",index=None)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()

model.fit(x_train,y_train)

print("The accuracy of the model is :",model.score(x_test,y_test)*100)
pred=model.predict(dft)
```

```python
sub["Survived"]=pred
sub.to_csv("sub dec",index=None)
```

```python
import catboost as ctb
model=ctb.CatBoostClassifier()
model.fit(x_train,y_train)

print("The accuracy of the model is :",model.score(x_test,y_test)*100)
pred=model.predict(dft)
sub["Survived"]=pred
sub.to_csv("sub cb",index=None)
```

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model.fit(x_train,y_train)

print("The accuracy of the model is :",model.score(x_test,y_test)*100)
pred=model.predict(dft)
sub["Survived"]=pred
sub.to_csv("sub nv",index=None)
```
